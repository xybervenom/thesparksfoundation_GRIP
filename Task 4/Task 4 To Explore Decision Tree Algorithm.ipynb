### Importing and Reading Data

```python
# import necessary modules  
import pandas  as pd 
import numpy as np
import matplotlib.pyplot as plt 
import matplotlib as mpl
import warnings
warnings.filterwarnings('ignore')
%matplotlib inline
```

```python
df=pd.read_csv("Iris.csv")
df.head()
```

```python
print(df.info())
```

#### Findings: 150 Rows, 6 Columns, 0 Null Values

```python
df.describe()
```

```python
print("Types of Species we are dealing with: ")
print(df['Species'].value_counts())
print("Data is Equally divided")
```

```python
# clean up the dataset to remove unnecessary columns (eg. REG)
df.drop(['Id'], axis=1, inplace=True)
df['Species'].replace(['Iris-versicolor','Iris-virginica','Iris-setosa'] , [0,1,2], inplace = True) #Categorizing
df.head()
```

## Splitting Data

```python
X = df.drop(labels='Species', axis=1) # Features
y = df.loc[:,'Species']  
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=1)
```

```python
print("Shape of Training Dataset", X_train.shape)
print("Shape of Test Dataset", X_test.shape)
```

## Decision Tree (Model 2)

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
clf=DecisionTreeClassifier()
clf.fit(X_train, y_train)
y_pred=clf.predict(X_test)
```

#### Confusion Matrix

```python
from sklearn.metrics import confusion_matrix,classification_report
cm = confusion_matrix(y_test, y_pred)
import itertools
def plot_confusion_matrix(cm, classes,title='Confusion matrix',cmap=plt.cm.Blues):
    plt.imshow(cm, interpolation='nearest', cmap=cmap)
    plt.title(title)
    plt.colorbar()
    tick_marks = np.arange(len(classes))
    plt.xticks(tick_marks, classes, rotation=45)
    plt.yticks(tick_marks, classes)
    fmt = 'd'
    thresh = cm.max() / 2.
    for i, j in itertools.product(range(cm.shape[0]), range(cm.shape[1])):
        plt.text(j, i, format(cm[i, j], fmt),horizontalalignment='center',color='white' if cm[i, j] > thresh else 'black')
    plt.tight_layout()
    plt.ylabel('True label')
    plt.xlabel('Predicted label')
    
class_names=np.array(['0','1','2'])
plot_confusion_matrix(cm,class_names)
print(classification_report(y_test,y_pred))
```

```python
from sklearn.metrics import accuracy_score
acc_ts=accuracy_score(y_pred, y_test)
print('Test Accuracy',acc_ts)
```
